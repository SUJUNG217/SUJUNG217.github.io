```python
#Import Libraries
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets
from sklearn.decomposition import PCA
from bokeh.io import push_notebook,show,output_notebook

import networkx as nx
```

```python
#Data Collection
iris=pd.read_csv("Iris.csv")
iris.head()
```

```python
#Matplotlib
iris["Species"].value_counts()
```

```python
#Scatter Plot
plt.figure()
plt.scatter(x=iris["SepalLengthCm"].values,y=iris["SepalWidthCm"].values)
```

```python
#Linear Plot
linear_data=np.array([1,2,3,4,5,6,7,8])
exponential_data=linear_data**2
plt.figure()
plt.plot(linear_data,'-o',exponential_data,'-*')
```

```python
#Bar Plot
plt.figure()
xvals=range(len(linear_data))
plt.bar(xvals,linear_data,width=0.3)
```

```python
from bokeh.plotting import figure,show,output_file

colormap={"Iris-setosa":"red","Iris-versicolor":"green","Iris-virginica":"blue"}
colors=[colormap[x] for x in iris["Species"]]

p=figure(title="Iris Morphology")
p.xaxis.axis_label="Petal Length"
p.yaxis.axis_label="Petal Width"

p.circle(iris["PetalLengthCm"],iris["PetalWidthCm"],color=colors,fill_alpha=0.2,size=10)

output_file("iris.html",title="iris.py example")

show(p)
```
